# Visualización preliminar de los datos

```python
# Importar las bibliotecas necesarias
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Cargar los datos (supongamos que tu conjunto de datos se llama 'datos.csv')
df = pd.read_excel('/content/MCD BD RENOV.xlsx')

```

```python
# Gráfica de barras para la Variable5
sns.countplot(x='Renovado', data=df)
plt.title('Renovaciones')
plt.show()

```

Hay 612 Renovaciones (1) y 315 que no renovaron (0)

```python
# Calcular la suma total por territorio
suma_total_por_territorio = df.groupby('Territorio')['Renovado'].sum().reset_index()
suma_total_por_territorio_ceros = (df == 0).sum()
# Gráfico de barras con etiquetas
plt.figure(figsize=(8, 6))
ax = sns.barplot(x='Territorio', y='Renovado', data=suma_total_por_territorio)

# Agregar etiquetas a las barras
for p in ax.patches:
    ax.annotate(f'{p.get_height():.0f}', (p.get_x() + p.get_width() / 2., p.get_height()),
                ha='center', va='center', xytext=(0, 10), textcoords='offset points', fontsize=10)

plt.title('Suma Total de Renovaciones por Territorio')
plt.xlabel('Territorio')
plt.ylabel('Suma Total de Renovaciones')
plt.show()
```

En el caso por territorios se tiene un porcentaje de renovación de 22.22% en Centro, 22.44% en Norte y 21.36% en Sur, el territorio con menor cantidad de renovaciones.

```python
# Crear un gráfico de bigotes para las cuatro variables
plt.figure(figsize=(10, 6))
sns.boxplot(data=df[['Ciclo','Integrantes','Renovado']], width=0.5)
plt.title('Gráfico de Bigotes')
plt.show()
```

```python
# Crear un gráfico de bigotes para las cuatro variables
plt.figure(figsize=(10, 6))
sns.boxplot(data=df['Monto'], width=0.5)
plt.title('Gráfico de Bigotes para Monto')
plt.show()
```

En las gráficas anteriores se muestran barras de bigotes las cuales podemos ver los outliers, no se eliminarán porque se perdería información importante por la cantidad de datos y magnitud que implican.
