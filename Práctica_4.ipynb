```python
# Importar las bibliotecas necesarias
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Cargar los datos (supongamos que tu conjunto de datos se llama 'datos.csv')
datos = pd.read_excel('/content/MCD BD RENOV.xlsx')

```

```python
# Calcular la matriz de correlaci칩n
correlation_matrix = datos.corr()

# Visualizar la matriz de correlaci칩n utilizando seaborn
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=.5)
plt.title('Matriz de Correlaci칩n')
plt.show()
```

Ciclo - Monto: estas variables muestran una correlaci칩n moderada de 0.50, indica que a mayor cantidad de ciclo hay un mayor monto otorgado.

Ciclo - Integrantes: se muestra una correlaci칩n d칠bil de 0.29 indica que a mayor cantidad de ciclos, ligeramente aumentar치 tambi칠n los integrantes.

Ciclo - Renovado: la correlaci칩n de 0.20 indica que es d칠bil por lo tanto, a mayor cantidad de ciclos, ligeremente mayor cantidad de renocaciones.

Integrantes - monto: muestra una correlaci칩n fuerte de 0.68 lo que indica que a mayor cantidad de integrantes, mayor cantidad de monto otorgado.

Integrantes - Renovado: muestra una correlaci칩n muy baja lo que indica que a mayor cantidad de integrantes, es marginalmente mayor la cantidad de renovaciones.

Monto - Renovado: muestra tambi칠n una correlaci칩n moderadamente baja de 0.17, indica que a mayor monto, ligeramente mayor cantidad de renovaciones
