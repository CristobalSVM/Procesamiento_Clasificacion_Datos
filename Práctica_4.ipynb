```python
# Importar las bibliotecas necesarias
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Cargar los datos (supongamos que tu conjunto de datos se llama 'datos.csv')
datos = pd.read_excel('/content/MCD BD RENOV.xlsx')

```

```python
# Calcular la matriz de correlación
correlation_matrix = datos.corr()

# Visualizar la matriz de correlación utilizando seaborn
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=.5)
plt.title('Matriz de Correlación')
plt.show()
```

Ciclo - Monto: estas variables muestran una correlación moderada de 0.50, indica que a mayor cantidad de ciclo hay un mayor monto otorgado.

Ciclo - Integrantes: se muestra una correlación débil de 0.29 indica que a mayor cantidad de ciclos, ligeramente aumentará también los integrantes.

Ciclo - Renovado: la correlación de 0.20 indica que es débil por lo tanto, a mayor cantidad de ciclos, ligeremente mayor cantidad de renocaciones.

Integrantes - monto: muestra una correlación fuerte de 0.68 lo que indica que a mayor cantidad de integrantes, mayor cantidad de monto otorgado.

Integrantes - Renovado: muestra una correlación muy baja lo que indica que a mayor cantidad de integrantes, es marginalmente mayor la cantidad de renovaciones.

Monto - Renovado: muestra también una correlación moderadamente baja de 0.17, indica que a mayor monto, ligeramente mayor cantidad de renovaciones
